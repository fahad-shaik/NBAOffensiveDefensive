```python
from nba_api.stats.endpoints import leaguedashplayerstats
# import json
import pandas as pd
```

```python
lsm_json = leaguedashplayerstats.LeagueDashPlayerStats(
    last_n_games = '0',
    measure_type_detailed_defense = 'Advanced',
    month = '0',
    opponent_team_id = '0',
    po_round_nullable = '0',
    pace_adjust = 'N',
    per_mode_detailed = 'PerGame',
    period = '0',
    league_id_nullable = '00', 
    season_type_all_star = 'Regular Season',
    team_id_nullable = '0',
    two_way_nullable = '0',
    season = '2019-20')
```

```python
lsm_json
```

```python
lsm_json = lsm_json.get_normalized_dict()
```

```python
lsm_json = lsm_json['LeagueDashPlayerStats']
```

```python
df = pd.DataFrame(lsm_json)
```

```python
df.to_csv("GreatData.csv")
```
